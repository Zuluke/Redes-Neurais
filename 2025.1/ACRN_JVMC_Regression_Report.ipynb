### (e) SVM com Kernel
Implementei um SVM Regressor (SVR) de tal forma que:
- Kernel do SVR  RBF;
    - Kernel Gaussiano: $\mathbf{K}(x, x') = exp(-\gamma * ||x - x'||^2)$
        - $\gamma$: controla a "largura";
        - $||x - x'||^2$:  a distncia euclidiana ao quadrado.
    - Com essa Tranformao de Ncleo, conseguimos mapear o problema para um espao de dimenso superior e fazer a regresso
- gamma: Controla o alcance da influncia de cada ponto de treinamento
    - Valores baixos: deciso mais suave (considera pontos mais distantes);
    - Valores altos: deciso mais complexa (considera pontos mais prximos);

### Consideraes Finais

1. O MLP-BP apresentou o melhor desempenho geral, com menor MSE e maior R no conjunto de teste. Isso era esperado, pois redes neurais multicamadas so aproximadores universais de funes. O principal destaque vai para a rpida convergncia do modelo que com menos robustez que os outros modelos (converge com menos de 1000 iteraes e possui apenas 20 neurnios na camada oculta), consegue ter um desempenho superior aos outros dois. Contudo, conseguiramos ajustar os outros modelos para que desempenhassem de forma parecida (ponto 4. deste breve relatrio), porm aumentando suas complexidades.

2. O RBF teve desempenho intermedirio. Ele funciona bem para problemas com caractersticas radialmente simtricas, mas pode ter dificuldade com funes mais complexas.

3. O SVM com kernel RBF teve o "pior" desempenho neste problema. SVMs so poderosos para classificao binria, mas podem ser menos eficientes para problemas de regresso complexos
como este.

4. Pelos desempenhos nas etapas de validao, os modelos no me parecem ter perdido a capacidade de generalizao.

5. H modos de aumentar as respectivas acurcias (aqui, vistas pelo score $R^2$). O problema passa a ser analisar se no h overfitting dos modelos ao aumentar seus desempenhos:
   - MLP-BP: Aumentar o nmero de neurnios na camada escondida
   - RBF: Aumentar o nmero de centros na regio de deciso
   - SVM com Kernel RBF: Aumentar o parmetro gamma que controla o alcance da influncia dos pontos no problema

6. A funo proposta parece ser melhor aproximada pela arquitetura do MLP-BP.
